```python
# Import dependencies 
import matplotlib.pyplot as plt 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot 
plt.plot(x_axis, y_axis)
```

```python
# OO method 

fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color="blue", linewidth=2, label = 'Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")

plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# OO approach 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label = "Boston", linewidth=2, color="green", marker="D")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,45)
ax.legend()
ax.grid()
```

```python
# BAR CHART PRACTICE 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the bar plot 
plt.bar(x_axis, y_axis, label="Boston")

# Create labels 
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.legend()

```

```python
# Create a horizontal bar chart 
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# OO Approach ax.plt() 

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.
```

```python
# SCATTER PLOTS
```

```python
# plt.plot(x_axis, y_axis, 'o')
# OR
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
plt.legend()
plt.xlabel("Fare($)")
plt.xlim(0, 45)
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Make the dots larger by iterating through a list 
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

# Replot with larger dots
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use list comprehension instead of 3 lines of a for loop 
# format: new_list = [expression FOR item IN list IF conditional]

plt.scatter(x_axis, y_axis_larger, s=[i*3 for i in y_axis])
```

```python
# OO Approach for Scatter Plots 
```

```python
fig, ax = plt.subplots()
larger_points = [i*5 for i in y_axis]
ax.scatter(y_axis, x_axis, s=larger_points, c="skyblue", label= "Boston", alpha=.8, linewidths=2, edgecolors="black")
ax.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
# PIE CHARTS 
```

```python
plt.figure()
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, explode=explode_values, colors=colors, autopct="%.1f%%")
```

```python
# OO METHOD
fig, ax = plt.subplots(figsize=(8,8))
explode=(0, 0, .2, 0, 0, 0, .2, 0, 0, 0, 0, 0)
ax.pie(y_axis, labels=x_axis, autopct=("%.1f%%"), startangle=83, shadow=True, explode=explode, counterclock=False )
plt.show()
```
